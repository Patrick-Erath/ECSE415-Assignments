### Data Acquisition


```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Setting image_read variables.
categories = {
    0 : "Group",
    1 : "Angela",
    2 : "Kevin",
    3 : "Oscar",
}
folder = "./data/"

imgs = []
#Reading images
for idx in range(len(categories)):
    imgType = categories[idx] + "/"
    imgPath = os.listdir(folder+imgType)
    #Read images in subfolder
    
    for imgFileName in os.listdir(folder+imgType):
        imgs.append(cv2.cvtColor(cv2.imread(folder+imgType+imgFileName), cv2.COLOR_BGR2RGB))
    
#Displaying images
plt.figure(figsize=(20,20))
plt.subplot(141)
plt.imshow(imgs[0])
plt.title("Input Image #1"), plt.xticks([]), plt.yticks([])
plt.subplot(142)
plt.imshow(imgs[0])
plt.title("Input Image #2"), plt.xticks([]), plt.yticks([])
plt.subplot(143)
plt.imshow(imgs[0],cmap = "gray")
plt.title("Depth Map (Truth)"), plt.xticks([]), plt.yticks([])
plt.subplot(144)
plt.imshow(imgs[0],cmap = "gray")
plt.title("Depth Map (Truth)"), plt.xticks([]), plt.yticks([])
plt.show()
```
