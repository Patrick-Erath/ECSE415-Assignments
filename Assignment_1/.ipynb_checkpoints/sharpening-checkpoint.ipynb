```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Create function to sharpen an image following forumla in assignment instructions
def sharpen(cleanImg, blurredImg):
    return 2*cleanImg - blurredImg
```

```python
#Q2.1
#Box sharpening
kernel = np.ones((5,5), np.float32)/(5*5)
rice = cv2.imread('rice.png',cv2.IMREAD_COLOR);
rice_BoxBlur = cv2.filter2D(rice, -1, kernel)
rice_box_sharp = sharpen(rice,rice_BoxBlur)


#Gaussian sharpening
rice = cv2.imread('rice.png',cv2.IMREAD_COLOR);
rice_GaussianBlur = cv2.GaussianBlur(rice,(5,5),sigmaX = 2, sigmaY = 2)
rice_gaussian_sharp = sharpen(rice,rice_GaussianBlur)


rice_plt = cv2.cvtColor(rice, cv2.COLOR_BGR2RGB)
rice_box_sharp_plt = cv2.cvtColor(rice_box_sharp, cv2.COLOR_BGR2RGB)
rice_gaussian_sharp_plt = cv2.cvtColor(rice_gaussian_sharp, cv2.COLOR_BGR2RGB)

plt.subplot(131), plt.imshow(rice_plt) # 1x5 subplot, 1 selected
plt.title("Rice Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(rice_box_sharp_plt) # 1x5 subplot, 2 selected
plt.title("Rice BoxSharp"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(rice_gaussian_sharp_plt) # 1x5 subplot, 4 selected
plt.title("Rice GaussSharp"), plt.xticks([]), plt.yticks([])
plt.show()
```

#Q2.2
To sharpen an image is to make edges more prevalent in the image in order to more easily identify the objects. If a better result is considered to be with the one with the more sharpened edge, then the BOX SHARPENING is expected to giver better result as the gaussian blur takes into consideration less the surrounding pixels than the box filter. If the better result is the one that looks the most realistic, than the GAUSSIAN SHARPENING is expected to give better result as the the edges contrast less with the background.
