```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load image
img = cv2.imread("capture.png",cv2.IMREAD_GRAYSCALE)
# create copy

# Read a template from one of the resistors
h , w = img.shape[0:2]
temp = np.zeros((h,w))
for x in range(h):
    for y in range(w):
        if(img[x][y] >= 160):
            temp[x][y] = 255


cv2.imwrite( "pineapple.jpg", temp );

# display images
plt.figure(figsize=(20,20))
plt.subplot(131), plt.imshow(img, cmap='gray')
plt.title("Input Image"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(temp, cmap = 'gray')
plt.title("Template Image"), plt.xticks([]), plt.yticks([])


plt.show()
```
